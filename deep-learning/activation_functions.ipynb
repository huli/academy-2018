### Activation functions

Here we simply plot different activation functions that can be used for neural networks.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define some functions
def sigmoid(z):
     return 1/(1+np.exp(-z))

def relu(z):
    return z * (z > 0)
```

```python
# Visualize some activation functions
sequence = np.arange(-5, 5, 0.01) 
plt.figure(figsize=(15,12))
plt.plot(sequence, np.tanh(sequence), linewidth=2)
plt.plot(sequence, sigmoid(sequence), linewidth=2)
plt.plot(sequence, relu(sequence), linewidth=2)
plt.plot(sequence, sequence, linewidth=1, linestyle='--')
plt.ylim([-1.5, 1.5])
plt.grid()
plt.legend(['tanh','sigmoid', 'ReLU', 'linear'], fontsize=15)
plt.show()
```
